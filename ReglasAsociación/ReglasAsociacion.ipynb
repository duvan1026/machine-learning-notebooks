## <font color='blue'> Introducción </font>

Las reglas de asociación se utilizan como una técnica de minería de datos para descubrir hechos que ocurren en común dentro de un determinado conjunto de datos. Pueden ayudar a descubrir relaciones entre variables en grandes conjuntos de datos. 

Uno de los trabajos más descriptivos de las reglas de asociación fue presentado por Agrawal et al., en el que indicaban que las reglas de asociación permitían descubrir las relaciones entre datos obtenidos a gran escala de los sistemas de terminales de punto de venta de los supermercados. Se contaría con una colección de artículos (ítems) que un cliente compra en una misma transacción.

* Problema: identificar el conjunto de ítems que son adquiridos en conjunto.

* Intentar identificar reglas de la forma:

                                        ${pan,azúcar} \Rightarrow {leche}$

¿Qué indicaría? Pues en este contexto de las ventas de un supermercado, indicaría que un consumidor que compra pan y azúcar a la vez, es probable que compre también leche. Esta información es muy últil para utilizar en la toma de decisiones sobre marketing, por ejemplo para lanzar un producto nuevo, o para promoción sobre el precio, o para ubicación del producto. 

Este es un ejemplo típico de aplicación de las reglas de asociación al análisis de la cesta de la compra, pero en general, se pueden aplicar a muchas otras áreas, o incluso para web mining que se estudió en la clase anterior.

## <font color='blue'>  Relaciones entre atributos </font>

### Correlación

#### Definición

Las correlaciones son un capítulo de la estadística en el que con distintos coeficientes matemáticos-estadísticos se trata de resolver el problema de la intensidad y sentido de la relación entre datos pertenecientes a dos variables o atributos.

#### Tipos de correlación y coeficientes de correlación

**Dos variables**

* **Variables continuas**:
    * Dos variables continuas (correlación lineal simple): Coeficiente de correlación $r$ de Pearson.
    * Dos variables continuas (correlación curvilínea): $\eta$ o razón de correlación.
    * Si una variable es continua y la otra es continua pero se encuentra dicotomizada (correlación biserial): Coeficiente $r_{b^{\blacksquare}}$
    * Si una variable es continua y la otra dicotómica por naturaleza (correlación biserial puntual): Coeficiente $r_{bp^{\blacksquare}}$
    * Si ambas son continuas, pero están dicotomizadas (correlación tetracórica): Coeficiente $r_{t^{\blacksquare}}$
* **Variables dicotómicas por naturaleza (atributos)**:
    * Las dos variables son dicotómicas por naturaleza: Coeficiente de correlación $\phi$.
    * Si las dos variables se encuentran divididas en dos o más categorías: test $\chi{}^{2}$ y coeficiente $C$ de contingencia.
* **Variables ordinales (Correlación ordinal)**:
    * Coeficiente $\tau$ de Kendall y $\rho$ de Spearman.

**Más de dos variables**

* **Variables continuas**:
    * Correlación múltiple.
    * Correlación parcial.
* **Variables discretas**:
    * Correlación múltiple.
    * Correlación parcial.
    * Análisis multivariado.

#### Coeficientes de correlación lineal r

<img src="../Figuras/Correlacion.png" alt="Figura Correlación" width="700"/>

**Ecuaciones**
$$ r = \frac{S_{xy}}{s_{x}\cdot{}s_{y}}$$
$$ s^{2} = \frac{1}{n - 1}\cdot{}\sum_{i-1}^{n}(x_{i} - \bar{x})^{2}$$
$$ S_{xy} = \frac{1}{n-1}\cdot{}\sum_{i-1}^{n}(x_{i} - \bar{x})\cdot{}(y_{i}-\bar{y}) $$

#### Propiedades

* -1 $ \leq{} r \leq{} $ 1
* Si $r = 1$ o $ r = -1 $ entonces hay correlación lineal perfecta.
* Si $r$ está próximo a 1 o a -1 entonces hay una asociación lineal fuerte entre ambas variables.
* Si r es cercano a 0 entonces hay una asociación lineal muy débil.
* r no varía cuando en las variables se realiza un cambio de escala o origen.

#### Consideraciones

* Se trata de una medida matemática que luego hay que interpretar.
* Que el grado de correlación sea próximo a 0 no indica la inexistencia de relación entre dos variables.

## <font color='blue'> Reglas de asociación  </font>


> **Una regla de asociación es una proposición probabilística sobre la ocurrencia de ciertos estados en un conjunto de datos.**



Las **Reglas de asociación** describen una relación de asociación entre los elementos de un conjunto de datos. Permite el descubrimiento de tendencias o patrones en grandes bases de datos con el
objetivo de guiar futuras decisiones. Se trata de una técnica ampliamente utilizada en Minería de Datos.



### Definiciones

**Itemset**: conjunto de uno o más ítems. En el ejemplo de la cesta de la compra serían los artículos. Ej.: {pan, leche}
  
**K-itemset**: Itemset con k elementos.
  
**Soporte o cobertura de un itemset**: fracción de las transacciones que contiene el itemset. Ej.: soporte({leche, azúcar}) = 2/4

**Itemset frecuente**: Itemset con soporte igual o superior a un umbral de soporte establecido por el usuario (MinSupp).

**Regla de asociación**: $ X \rightarrow{} Y$, siendo x e y itemsets. Ej.: <br> {pañales} $\rightarrow$ {cerveza} <br> {pan,azúcar} $\rightarrow$ {leche}

### Medidas de evaluación de las reglas de asociación

**Soporte o cobertura de un itemset**: **soporte (A)** <br>
    Fracción de las transacciones que contienen a A.  <br>
    Se usa también el porcentaje.
    
**Confianza o precisión de la regla**: **confianza (A $\Rightarrow$ B) = P(A$\,|\,$B) = soporte (A $\cup$ B)** <br>
    Fracción de las transacciones en las que aparece A que también incluyen a B. La confianza mide con qué frecuencia aparece B en las transacciones que incluyen A. <br>
Porcentaje de veces que la regla se cumple cuando se puede aplicar.

 
**Las reglas de asociación:**
    
* **Relacionan conjuntos de atributos categóricos entre sí.**
* Siguen reglas del tipo: "**Si** A, B, ... **entonces** C, D, ...", donde A, B, ... forman **el antecedente** y C, D, ... componen **el consecuente**.
    
* Se puede medir la **calidad de una regla**
    $$ soporte(A \Rightarrow B) = soporte(A\cup{}B) = P(A U B)$$
    $$confianza(A \Rightarrow B) = P(B\,|\,A) = \frac{soporte(A\cup{}B)}{soporte(A)}$$

**Interpretación:**

* En una regla con baja confianza es probable que no exista relación entre antecedente y consecuente.
* ¿$A \Rightarrow B$ es igual a $B \Rightarrow A$? Tienen el mismo soporte pero distinta confianza.

**Objetivo:**
> Encontrar reglas de asociación con valores de soporte y confianza altos.
  Los umbrales _Minsup_ y _Minconf_ son definidos por el usuario.

### Algoritmo de aprendizaje básico

#### Fases

1. Extracción  de los conjuntos de ítems (pares atributo-valor) que cumplan con la cobertura (soporte) requerida (Algoritmo Apriori).
2. Generación de las reglas que superen un nivel de confianza mínimo a partir de esos conjuntos.

**Algoritmo de aprendizaje Apriori**
* Algoritmo de aprendizaje para obtener conjuntos de ítems frecuentes para reglas de asociación Booleanas.
* Usa conocimiento previo de las propiedades de los conjuntos de ítems frecuentes.

> **Principio**: un conjunto de ítems formado por X ítems es frecuente si y solo si cada uno de los X ítems es frecuente por sí solo.

> **Proceso**: genera todos los conjuntos de ítems con un elemento. Usa estos para generar los de dos elementos, y así sucesivamente.

**Algoritmo** <br>
**Input**: Database, D, of transactions; minimum support threshold, min_sup. <br>
**Output**: L, frequent itemsets in D. <br>
**Method**: <br>

1. $L_{1}$ = find_frequent_1-itemsets(D);
2. **for** ($k = 2;L_{k-1} \neq{} \phi; k++$)
3. &nbsp;&nbsp;&nbsp;&nbsp;&nbsp; $C_k$ = apriori_gen($L_{k-1}$, min_sup);
4. &nbsp;&nbsp;&nbsp;&nbsp;&nbsp; **for each** transaction $t \in D$ { // scan D for counts
5. &nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp; $C_{t} = subset(C_k, t);$ // get the subsets of t that are candidates
6. &nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp; **for each** candidate $c \in C_t$
7. &nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp; $c.count++$;
8. &nbsp;&nbsp;&nbsp;&nbsp;&nbsp; }
9. &nbsp;&nbsp;&nbsp;&nbsp;&nbsp; $L_{k}$ = {$c \in C_{k} \,|\, c.count \geq$ min_sup}
10. }
11. return $L = \cup_{k}L_k$

procedure apriori_gen($L_{k-1}:$ frequent $(k-1)$-itemsets; *min_sup*: minimum support)
1. &nbsp;&nbsp;&nbsp;&nbsp;&nbsp; **for each** itemset $l_1 \in L_{k-1}$
2. &nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp; **for each** itemset $l_2 \in L_{k-1}$
3. &nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp; **if**($l_{1}[1] = l_{2}[1]) \wedge (l_{1}[2] = l_{2}[2]) \wedge \dots \wedge  (l_{1}[k-2] = l_{2}[k-2]) \wedge{} (l_{1}[k - 1] < l_{2}[k-1])$ **then** {
4. &nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp; $c = l_{1} \bowtie{} l_{2}$; // join step_ generate candidates
5. &nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp; **if** has_infrequent_subset($c, L_{k-1}$) **then**
6. &nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp; **delete** $c$; // prune step: remove unfruitful candidate
7. &nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp; **else add** $c$ **to** $C_k$;
8. &nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp; }
9. &nbsp;&nbsp;&nbsp;&nbsp;&nbsp; **return** $C_k$;

procedure has_infrequent_subset($c$: candidate $k$-itemset; $L_{k-1}$: frequent ($k-1)$-itemsets); // use prior knowledge
1. &nbsp;&nbsp;&nbsp;&nbsp;&nbsp; **for each** ($k-1)$-subset $s$ **of** $c$
2. &nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp; **if** s $\notin L_{k-1}$ **then**
3. &nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp; **return** TRUE;
4. &nbsp;&nbsp;&nbsp;&nbsp;&nbsp; **return** FALSE;

**Fase 2**

> **Propiedad**: si una conjunción de consecuentes de una regla cumple con los niveles mínimos de soporte y confianza, sus subconjuntos (consecuentes) también los cumplen. Por el contrario, si algún ítem no los cumple, no tiene caso considerar sus superconjuntos.

> **Proceso**: genera reglas con un solo consecuente y, a partir de ellas, dos consecuentes, tres consecuentes, ...

El proceso para generar las reglas a partir de los ítems es el siguiente:

* Para cada conjunto $l$ de items, genera todos sus subconjuntos.
* Para cada subconjunto $s \subset{} l$, genera una regla: $s \Rightarrow (l - s)$ si:

$$ \frac{soporte (l)}{soporte (s)} \geq nivel_{confianza}$$

**Ejemplo 2**

Continuando con el ejemplo, a partir de $L_{3}$:


* $l$ = {$I_{2}$,$I_{3}$,$I_{4}$}
* Los subconjuntos no vacios $s$ de $l$ son:
    * {$I_{2}$,$I_{3}$}
    * {$I_{2}$,$I_{4}$}
    * {$I_{3}$,$I_{4}$}
    * {$I_{2}$}
    * {$I_{3}$}
    * {$I_{4}$}
* Las reglas de asociación obtenidas son:
    * SI $I_{2}$ Y $I_{3}$ ENTONCES $I_{4}$ &nbsp;&nbsp; *Confianza* = 2/2 = 100%
    * SI $I_{2}$ Y $I_{4}$ ENTONCES $I_{3}$ &nbsp;&nbsp; *Confianza* = 2/2 = 100%
    * SI $I_{3}$ Y $I_{4}$ ENTONCES $I_{2}$ &nbsp;&nbsp; *Confianza* = 2/3 = 67%
    * SI $I_{2}$ ENTONCES $I_{3}$ Y $I_{4}$ &nbsp;&nbsp; *Confianza* = 2/3 = 67%
    * SI $I_{3}$ ENTONCES $I_{2}$ Y $I_{4}$ &nbsp;&nbsp; *Confianza* = 2/3 = 67%
    * SI $I_{4}$ ENTONCES $I_{2}$ Y $I_{3}$ &nbsp;&nbsp; *Confianza* = 2/3 = 67%
* Si el nivel de confianza mínimo es del 50%, todas las reglas son admitidas. Pero si se establece el umbral mínimo al 70%, solo las dos primeras serían aceptadas.


**Implementación de este ejemplo**

```python
from collections import Counter

from itertools import combinations

data = [
        ['T100',['I1','I2','I5']],
        ['T200',['I2','I3','I4']],
        ['T300',['I3','I4']],
        ['T400',['I1','I2','I3','I4']],
        ]

init = []

for i in data:
    for q in i[1]:
        if(q not in init):
            init.append(q)
            
init = sorted(init)
print(init)

sp = 0.4
s = int(sp*len(init))

c = Counter()

for i in init:
    for d in data:
        if(i in d[1]):
            c[i]+=1
            
print("C1:")

for i in c:
    print(str([i])+": "+str(c[i]))
    
print()

l = Counter()
for i in c:
    if(c[i] >= s):
        l[frozenset([i])]+=c[i]
        
print("L1:")
for i in l:
    print(str(list(i))+": "+str(l[i]))
print()

pl = l
pos = 1
for count in range (2,1000):
    nc = set()
    temp = list(l)
    for i in range(0,len(temp)):
        for j in range(i+1,len(temp)):
            t = temp[i].union(temp[j])
            if(len(t) == count):
                nc.add(temp[i].union(temp[j]))
    nc = list(nc)
    c = Counter()
    for i in nc:
        c[i] = 0
        for q in data:
            temp = set(q[1])
            if(i.issubset(temp)):
                c[i]+=1
    print("C"+str(count)+":")
    for i in c:
        print(str(list(i))+": "+str(c[i]))
    print()
    l = Counter()
    for i in c:
        if(c[i] >= s):
            l[i]+=c[i]
    print("L"+str(count)+":")
    for i in l:
        print(str(list(i))+": "+str(l[i]))
    print()
    if(len(l) == 0):
        break
    pl = l
    pos = count
    
print("Resultado: ")
print("L"+str(pos)+":")
for i in pl:
    print(str(list(i))+": "+str(pl[i]))
print()

for l in pl:
    c = [frozenset(q) for q in combinations(l,len(l)-1)]
    mmax = 0
    for a in c:
        b = l-a
        ab = l
        sab = 0
        sa = 0
        sb = 0
        for q in data:
            temp = set(q[1])
            if(a.issubset(temp)):
                sa+=1
            if(b.issubset(temp)):
                sb+=1
            if(ab.issubset(temp)):
                sab+=1
        temp = sab/sa*100
        if(temp > mmax):
            mmax = temp
        temp = sab/sb*100
        if(temp > mmax):
            mmax = temp
        print(str(list(a))+" -> "+str(list(b))+" = "+str(sab/sa*100)+"%")
        print(str(list(b))+" -> "+str(list(a))+" = "+str(sab/sb*100)+"%")
    curr = 1
    print("Elección:", end=' ')
    for a in c:
        b = l-a
        ab = l
        sab = 0
        sa = 0
        sb = 0
        for q in data:
            temp = set(q[1])
            if(a.issubset(temp)):
                sa+=1
            if(b.issubset(temp)):
                sb+=1
            if(ab.issubset(temp)):
                sab+=1
        temp = sab/sa*100
        if(temp == mmax):
            print(curr, end = ' ')
        curr += 1
        temp = sab/sb*100
        if(temp == mmax):
            print(curr, end = ' ')
        curr += 1
    print()
    print()
```

### Consideraciones

#### Reglas redundantes

Se considera que existe redundancia cuando en el antecedente de una regla aparece el antecedente de otra con algunos items más. Es decir, existe una regla cuyo antecedente es un subconjunto del antecedente de otra regla. En este caso, la regla más genérica se considera redundante, no aporta información adicional. Por tanto, se considera que la regla $A \Rightarrow B$ es redundante si un subconjunto A' de A aparece como antecedente de otra regla $A' \Rightarrow B$c uyo soporte es mayor.

$A \Rightarrow B$ es redundante si existe un subset A’ tal que: conf($A' \Rightarrow B$) >= conf($A \Rightarrow B$)

#### Ausencia de ítems

* Problema del algoritmo: no se permite considerar reglas que contemplen la ausencia de un ítem.
    * SI $I_{3}$ Y $I_{4}$ ENTONCES $\neg{}I_{5}$ &nbsp; 
* Solución: crear dos ítems a partir de cada ítem original que tomen los valores verdadero y falso.
    * $I_{5} = 1$, $I_{6} = 0$

#### Atributos no binarios

* Problema del algoritmo: genera reglas de asociación booleanas, es decir, trabaja con atributos booleanos.
* Solución: considerar como ítem cada posible valor diferente de un determinado atributo (generalización de la solución al problema anterior).
    * nivel_vision = buena, nivel_vision = regular, nivel_vision = mala

#### Dependencias entre atributos

La correlación entre dos conjuntos de ítems A y B se define:
$$ corr_{A,B} = \frac{P(A\cap{}B)}{P(A)P(B)}$$
* $<$ 1 Ocurrencia de uno decrece la del otro 
* $=$ 1 son independientes
* $>$ 1 la ocurrencia de uno favorece la del otro

**Ejemplo3 (Dependencias)**

| Tipo Pan | PAN INTEGRAL | ¬(PAN INTEGRAL) | SUMA |
|---------------|:------------:|:---------------:|:----:|
| **PAN BLANCO** | 20 | 60 | 80 |
| **¬(PAN BLANCO)** | 10 | 10 | 20 |
| **SUMA** | 30 | 70 | 100 |

<br>

SI *Pan Integral* ENTONCES *Pan Blanco* &nbsp;&nbsp; *cob.* = 20% &nbsp;&nbsp; *conf.* = 66%
$$corr_{Pan Integral, Pan Blanco} = \frac{P(Pan Blanco \cap Pan Integral)}{P(Pan Blanco)P(Pan Integral)} = \frac{0,2}{0,8\cdot0,3} = 0,833$$

Existe una dependencia negativa entre comprar pan integral y pan blanco (comprar pan integral disminuye la probabilidad de que se compre pan blanco).


<br>

<br>
SI ¬ *Pan Blanco* ENTONCES *Pan Integral* &nbsp;&nbsp; *cob.* = 10% &nbsp;&nbsp; *conf.* = 50%
$$corr_{¬Pan Blanco,Pan Integral} = \frac{P(¬Pan Blanco \cap Pan Integral)}{P(¬Pan Blanco)P(Pan Integral)} = \frac{0,1}{0,2\cdot0,3} = 1,66$$

Existe una fuerte dependencia positiva entre no comprar pan blanco y comprar pan integral (mayor que la regla anterior, a pesar de tener menor confianza).

$$= 0,667 + 0,286 + 2,667 + 1,143 = 4,76$$

El valor de la $\chi{}^{2}$ para el 95% de la significancia y un grado de libertad es 3,85 $<$ 4,76 $\Rightarrow$ rechazamos la hipótesis de independencia con un 95% de certeza.

### Importar las librerías

El primer paso es importar las librerías requeridas.

```python
# Tratamiento de datos
# ------------------------------------------------------------------------------
import pandas as pd  

# Algoritmo
# ------------------------------------------------------------------------------
from apyori import apriori 

# Para su instalación 
#!pip install apyori
```

### Importar el conjunto de datos

```python
dataset = pd.read_csv('DataSets/shopBask.csv',header=None)   
dataset.head(10)
```

Si observa detenidamente los datos, cada fila corresponde a una transacción y cada columna corresponde a un artículo comprado en esa transacción específica. NaN nos dice que el artículo representado por la columna no fue comprado en esa transacción específica.

Por defecto, la función pd.read_csv trata la primera fila como encabezado. En este conjunto de datos no hay fila de encabezado. Por lo tanto, se agrega la opción header=None a la función pd.read_csv.

El algoritmo Apriori se usa para averiguar qué artículos se venden juntos comúnmente, por lo que esta información se puede considerar para colocar los artículos relacionados juntos o publicitarlos para aumentar las ganancias.

### Preprocesar los datos

El algoritmo Apriori requiere que el conjunto de datos tenga la forma de una lista de listas. El conjunto de datos completo es una lista grande y cada transacción en el conjunto de datos es una lista interna dentro de la lista externa. Actualmente, los datos tienen la forma de un dataframe de pandas. Para convertir el dataframe de pandas en una lista de listas, ejecute el siguiente script:

```python
rows = len(dataset.index)
print(rows)
cols = len(dataset.columns)
print(cols)
records = []  
for i in range(0, rows):  
    records.append([str(dataset.values[i,j]) for j in range(0, cols)])
records
```

Otra alternativa es convertir el marco de datos en una matriz 0/1 en la que cada fila es una transacción y cada columna es uno de los elementos posibles.

### Aplicar algoritmo Apriori

El siguiente paso es aplicar el algoritmo Apriori sobre el conjunto de datos. Para hacerlo, podemos usar la clase _apriori_ que importamos de la librería apyori.

La clase _apriori_ requiere algunos valores de parámetros para funcionar. El primer parámetro es la lista de lista de la que desea extraer reglas. El segundo parámetro es el parámetro *min_support*. Este parámetro se utiliza para seleccionar los elementos con valores de soporte superiores al valor especificado por el parámetro. A continuación, el parámetro *min_confidence* filtra aquellas reglas que tienen una confianza superior al umbral de confianza especificado por el parámetro. De manera similar, el parámetro *min_lift* especifica el valor de elevación mínimo para las reglas preseleccionadas. Finalmente, el parámetro *min_length* especifica la cantidad mínima de elementos que desea en sus reglas.

Supongamos que queremos reglas sólo para aquellos artículos que se compran al menos 5 veces al día, o 7 x 5 = 35 veces en una semana, ya que el conjunto de datos es para un período de una semana. El soporte para esos artículos se puede calcular como 35/7500 = 0.0045. La confianza mínima para las reglas es 20% o 0.2. Igualmente, especificamos el valor de lift como 3 y finalmente *min_length* es 2 ya que queremos al menos dos productos en nuestras reglas. Estos valores en su mayoría se eligen arbitrariamente, por lo que puede jugar con estos valores y ver qué diferencia  provoca en las reglas.

Ejecute el siguiente script:

```python
rules = apriori(records, min_support=0.5, min_confidence=0.7)
rules
```

Podemos convertir las reglas encontradas por la clase _apriori_ en una lista, ya que es más fácil ver los resultados en este formulario.

```python
associationRules = list(rules)
associationRules
```

### Visualizar los resultados

Primero encontremos el número total de reglas extraídas por la clase _apriori_. Ejecute el siguiente script:

```python
print(len(associationRules))
```

Cada elemento corresponde a una regla.

Imprimamos un elemento de la lista associationRules para ver una de las reglas. Ejecute el siguiente script:

```python
print(associationRules[3])
```

El valor de soporte para la primera regla es 0.57. Este número se calcula dividiendo el número de transacciones que contienen 'soda' y 'juice' por el número total de transacciones.

El nivel de confianza de la regla es 0,79, lo que muestra que de todas las transacciones que contienen 'soda', el 79,9%, contienen 'juice'.

El _lift_ de 0,93 nos dice que 'juice' es 0,93 veces más probable a ser comprado por los clientes que compran 'soda', en comparación con la probabilidad predeterminada de venta de 'juice'.


Puede implementar un script para mostrar la regla, el soporte, la confianza y el _lift_ de cada regla de una manera más clara. O puede convertirlo a DataFrame:

```python
aprioridf = pd.DataFrame(associationRules)
print(aprioridf)
```
